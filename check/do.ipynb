```python
# Read crispor.csv
df = pd.read_csv('crispor.csv')
# Read my.csv
df2 = pd.read_csv('my.csv')
# Change DNA column to upper case.
df2['DNA'] = df2['DNA'].str.upper()
# If DNA in df is in DNA in df2 with the same #Id, df['my']=1, else=0.
df['my'] = None
for idx, row in df.iterrows():
    if row['DNA'] in df2[df2['#Id'] == row['#Id']]['DNA'].values:
        df.loc[idx, 'my'] = 1
    else:
        df.loc[idx, 'my'] = 0
df.to_csv('final.csv', index=False)
```
